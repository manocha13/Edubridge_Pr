```python
import matplotlib.pyplot as plt
```

```python
#scatterplot
import numpy as np
```

```python
x = np.arange(1,10)
y = np.arange(1,10)
```

```python
plt.scatter(x,y)
plt.plot(x,y)
plt.xlabel("X axis")
plt.title("Scatter plot")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.scatter(x,y,c="green",marker="o",s=100,alpha=1)
plt.plot(x,y,c="red",linestyle="dashdot",linewidth=2)
plt.xlabel("X axis")
plt.title("Scatter plot")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.scatter(x,y,c=['green','red','black','blue','yellow','purple','pink','hotpink','orange'],marker="o",s=100,alpha=1)
plt.plot(x,y,c="red",linestyle="dashdot",linewidth=2)
plt.xlabel("X axis")
plt.title("Scatter plot")
plt.ylabel("Y axis")
plt.show()
```

```python
colors = np.arange(1,10)
plt.scatter(x,y,c=colors,marker="o",s=100,alpha=1,cmap="Blues",edgecolor="red")
plt.plot(x,y,c="black",linestyle="dashdot",linewidth=2)
plt.xlabel("X axis")
plt.title("Scatter plot")
plt.ylabel("Y axis")
plt.show()
```

```python
z = np.arange(2,20,2)
colors = np.arange(1,10)
plt.scatter(x,y,c=colors,marker="o",s=100,alpha=1,cmap="Blues",edgecolor="red")
plt.plot(x,y,c="black",linestyle="dashdot",linewidth=2,label="X vs Y")

plt.scatter(x,z)
plt.plot(x,z,label="X vs Z")

plt.legend(bbox_to_anchor=(-0.1,0.3))
plt.xlabel("X axis")
plt.title("Scatter plot")
plt.grid()
plt.ylabel("Y axis")

for i in range(len(x)):
    plt.text(x[i]+0.2,y[i]-0.4,(x[i],y[i]))
plt.rcParams['figure.figsize'] = (10,5)
plt.show()
```

```python
z = np.arange(2,20,2)
colors = np.arange(1,10)

plt.subplot(2,1,1)
plt.scatter(x,y,c=colors,marker="o",s=100,alpha=1,cmap="Blues",edgecolor="red")
plt.plot(x,y,c="black",linestyle="dashdot",linewidth=2,label="X vs Y")

plt.subplot(2,1,2)
plt.scatter(x,z)
plt.plot(x,z,label="X vs Z")

plt.legend(bbox_to_anchor=(-0.1,0.3))
plt.xlabel("X axis")
plt.title("Scatter plot")
plt.grid()
plt.ylabel("Y axis")

for i in range(len(x)):
    plt.text(x[i]+0.2,y[i]-0.4,(x[i],y[i]))
plt.rcParams['figure.figsize'] = (5,5)
plt.show()
```

```python
z = np.arange(2,20,2)
colors = np.arange(1,10)

plt.subplot(1,2,1)
plt.scatter(x,y,c=colors,marker="o",s=100,alpha=1,cmap="Blues",edgecolor="red")
plt.xlabel("X axis")
plt.title("Scatter plot")
plt.grid()
plt.ylabel("Y axis")
#plt.legend()
plt.plot(x,y,c="black",linestyle="dashdot",linewidth=2)

plt.subplot(1,2,2)
plt.scatter(x,z)
plt.plot(x,z,label="X vs Z")

#plt.legend(bbox_to_anchor=(0.5,0.3))
plt.xlabel("X axis")
plt.title("Scatter plot")
plt.grid()
plt.ylabel("Y axis")

for i in range(len(x)):
    plt.text(x[i]+0.2,y[i]-0.4,(x[i],y[i]))
plt.rcParams['figure.figsize'] = (10,5)
plt.show()
```

```python
#bar charts
marks = np.random.randint(50,100,10)
name= ["Anshul","Balwinder","Bhakti","Sanjai","Sajal","Mohammad","Gayatri","Andrew","Saswat","Sharadha"]

colors = ["green","red","blue","orange","purple","pink","hotpink","black","white","yellow"]

ax = plt.axes()

ax.set_facecolor("maroon")
plt.xticks(rotation=360)
plt.bar(name,marks,color=colors,edgecolor="black",width=0.5)
plt.show()
```

```python
#stacked bar charts
marks = np.random.randint(50,100,10)
name= ["Anshul","Balwinder","Bhakti","Sanjai","Sajal","Mohammad","Gayatri","Andrew","Saswat","Sharadha"]
marks_2 = np.random.randint(50,100,10) 
colors = ["r","b","purple","hotpink","purple","pink","yellow","green","orange","yellow"]

colors_2 = ["r","b","r","b","r","b","green","orange","yellow","purple"]
ax = plt.axes()

ax.set_facecolor("maroon")
plt.xticks(rotation=360)
plt.bar(name,marks,color=colors,edgecolor="black",width=0.5)
plt.bar(name,marks_2,color=colors_2,edgecolor="black",width=0.5,bottom=marks)
plt.show()
```

```python
#bar charts
marks = np.random.randint(50,100,10)
name= ["Anshul","Balwinder","Bhakti","Sanjai","Sajal","Mohammad","Gayatri","Andrew","Saswat","Sharadha"]
marks_2 = np.random.randint(50,100,10) 
colors = ["r","b","purple","hotpink","purple","pink","yellow","green","orange","yellow"]

colors_2 = ["r","b","r","b","r","b","green","orange","yellow","purple"]

#position
bar1 = np.arange(len(marks))
bar2 = [i+0.2 for i in bar1]


ax = plt.axes()

ax.set_facecolor("maroon")
plt.xticks(rotation=360)
plt.bar(name,marks,color=colors,edgecolor="black",width=0.2,align='center')
plt.bar(name,marks_2,color=colors_2,edgecolor="black",width=0.2,align='edge')
plt.show()
#plt.bar()
```

```python
#data
IT = [10,15,16,12,14]
ECE = [13,15,12,16,17]
Fashion = [20,14,17,18,19]

#position
bar1 = np.arange(len(IT))
bar2 = [i+0.20 for i in bar1]
bar3 = [i+0.20 for i in bar2]

plt.bar(bar1,IT,width=0.2)
plt.bar(bar2,ECE,width=0.2)
plt.bar(bar3,Fashion,width=0.2)

plt.xticks([i+0.2 for i in range(len(IT))],["2015","2016","2017","2018","2019"])
#plt.legend(["IT","ECE","Fashion"],loc="best")
plt.legend(["IT","ECE","Fashion"],bbox_to_anchor=(0,0))
plt.show()
```

```python
#distribution of the data 
#cake#percentage
#pie charts

x = [25,25,25,25]
l = ["A","B","C","D"]
plt.pie(x,labels=l,counterclock=False,startangle=90)
plt.show()
```

```python
#distribution of the data 
#cake#percentage
#pie charts

x = [80,90,70,60]
l = ["East","Central","North","South"]
plt.pie(x,labels=l,counterclock=False,colors=["g","r","b","orange"],radius=1,pctdistance=0.8,wedgeprops = {'linewidth': 1.5,'edgecolor':"black"},autopct ="%0.2f%%",startangle=90,explode = [0.1,0.2,0,0],labeldistance=1.1,shadow="True")
plt.legend(bbox_to_anchor=(0.9,0.9),title="Region")
plt.axis("equal")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('winequality-red.csv')
```

```python
#boxplots
df.describe()
```

```python
y = np.arange(1,17)
plt.boxplot(df['fixed acidity'])
plt.xticks([1],['Fixed Acidity'])
plt.yticks(y)
plt.grid()
plt.show()
```

```python
var1 = df['fixed acidity']
var2 = df['volatile acidity']
var3 = df['citric acid']

data = [var1,var2,var3]

fig, ax=plt.subplots()
ax.boxplot(data,notch=True,patch_artist=True)
plt.xticks([1,2,3],['Fixed Acidity','volatile acidity','citric acid'])
plt.show()
```

```python
var1 = df['fixed acidity']
var2 = df['volatile acidity']
var3 = df['citric acid']

data = [var1,var2,var3]

fig, ax=plt.subplots()
ax.boxplot(data,vert=0,notch=True,patch_artist=True)
plt.xticks([1,2,3],['Fixed Acidity','volatile acidity','citric acid'])
plt.show()
```
